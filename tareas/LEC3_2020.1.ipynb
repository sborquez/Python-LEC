````
#Podemos además graficar la función acumulada, por ejemplo

x = np.linspace(norm.ppf(0.0001), norm.ppf(0.9999), 1000)

plt.plot(x, norm.cdf(x), 'b-', label='normal estandar')

x = np.linspace(norm.ppf(0.0001, loc=10, scale=5),
                norm.ppf(0.9999, loc=10, scale=5), 1000)

plt.plot(x, norm.cdf(x, loc=10, scale=5), 'r-', label='normal 10, 5')
plt.legend()
plt.show()
````

¿Cómo podemos generar muestras de una variable aleatoria, si solo se tiene disponible su **pdf** y **cdf**? Para casos como este, necesitamos además, un generador de números aleatorios distribuidos uniforme en el rango $[0,1]$. Con estos componentes y el siguente algoritmo, les permitiran generar muestras aleatorias desde cualquier distribución, con tan solo tener su **cdf**:


**Generador de muestras aleatorias de distribuciones continuas**

0. Elegir los parámetros para discretizar: $N$ número de divisiones, rango a considerar $[x_{min}, x_{max}]$, $M$ cantidad de muestras.

1. Definir:
  * $h$: Ancho de divisiones $\frac{x_{max} - x_{mix}}{N}$
  * $x_j$: $N+1$ puntos divisorios $x_j = x_{min} + jh$
2. Se evalua la cdf en los $N+1$ puntos. $\Phi_{Tj} = \Phi_{T}(x_j),\, \forall j \in [0, N]$
3. Tomar una muestra $m$ de una distribución $\text{Uniforme}(0,1)$.
4. Encontrar un $j$ tal que $$\Phi_{T(j-1)}  \lt m \leq \Phi_{Tj} $$
5. Obtener la muestra $x$ al interpolar: $$x = (1-\alpha)x_{j-1} + \alpha x_j$$. Donde: $$\alpha = \frac{m - \Phi_{T(j-1)}}{\Phi_{Tj} - \Phi_{T(j-1)}}$$
6. Repetir hasta obtener $M$ muestras.

Considere a la variable aleatoria $\Theta$, **independiente** de $T$, tal que:

$$\Theta  \sim \text{Uniform}([0, 2\pi])$$

$$\pi_{\Theta}(\theta) = \frac{1}{2\pi}$$

La covarianza entre $T$ y $\Theta$ es 0:


$$\text{cov}(T,\Theta) = E[T\Theta] - E[T]E[\Theta] = E[T]E[\Theta] - E[T]E[\Theta] = 0$$

Esto ocurre siempre, cuando dos variables aleatorias son independientes, se cumple: $E[AB] = E[A]E[B]$.

Se define las variables aleatorias $X$ e $Y$ como: 

$$X=R_0Tcos\Theta,\, Y=R_0Tsin\Theta$$

Con $R_0$ una constante.
  
Es fácil demostrar que $X$ e $Y$ no son independientes:


**Demostración:**

Obtenemos una expresión de $T$ y $\Theta$ en términos de $X$ e $Y$:

\begin{align}
X^2 &= R_0^2 T^2 \cos^2(\Theta)\\
Y^2 &= R_0^2 T^2 \sin^2(\Theta) \\
&\rightarrow X^2 + Y^2 =  R_0^2 T^2 (\cos^2(\Theta) + \sin^2(\Theta))\\
& \rightarrow T = \frac{(X^2 + Y^2)^{0.5}}{R_0}\\
& \rightarrow \Theta = \tan^{-1}(\frac{Y}{X})\end{align}


Con el Jacobiano:
\begin{align}
J(T,\Theta) &=\left. 
\right. 
\begin{vmatrix}
\frac{\partial T}{\partial X} & \frac{\partial T}{\partial Y} \\ 
\frac{\partial \Theta}{\partial X} & \frac{\partial \Theta}{\partial Y} 
\end{vmatrix}
= \frac{1}{R_0 (X^2+Y^2)^{\frac{1}{2}}} 
\end{align}


$X$ e $Y$ no son independentes, su distribución conjunta no puede escribirse como el producto de 2 dos funciones de cada variable.


\begin{align}
\pi(t,\theta) &=  \pi_T(t)*\pi_{\Theta}(\theta)\\
\Rightarrow \pi(x,y) &= \pi_T(\frac{(x^2+y^2)^{0.5}}{R_0})*\pi_{\Theta}(\tan^{-1}{\frac{y}{x}})* J(x,y) \\
&= \frac{1}{2} \frac{(x^2+y^2)}{R^2_0}e^{-\frac{(x^2+y^2)^{0.5}}{R_0}} * \frac{1}{2\pi} * \frac{1}{R^2_0 (x^2+y^2)^{0.5}}\\
&= \frac{1}{4\pi R^4_0} (x^2+y^2)^{0.5}e^{-\frac{(x^2+y^2)^{0.5}}{R_0}}   
\end{align}

## 4. Kernel Density Estimation

En el laboratorio anterior se discutió sobre el uso de las KDE. Estas nos permiten aproximar cualquier pdf de alguna distribución utilizando una función $\hat{f}$. Como de define a continuación:

$$\text{pdf}(x) \approx \hat{f}(x) = \frac{1}{nh}\sum_{x_k \in \text{ sample} } K(\frac{x-x_k}{h}) $$

Para definir $\hat{f}$, se necesita de tres componentes: una muestra de la distribución a aproximar; un kernel, una función simetrica y no negativa; y un _bandwidth_ o $h$ que nos permite controlar la aproximación.







````
gaussian_kernel = lambda xi: (1/np.sqrt(2*np.pi)) * np.exp((-1*xi**2)/2)
f_hat = lambda xi, sample, h, K: (1/(h*len(sample))) * np.sum(K((xi - sample)/h))
````
