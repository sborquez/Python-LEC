## 1. Combinando datasets

En la siguiente sección realizaremos un analisis más complejo, al incluir información geográfica para segregar la información. Para esto, el primero paso será implementar una función que obtenga la **cantidad de casos** de **cada país** para una **fecha dada**, incluyendo **información geodemográfica** de cada país.


* (<font color='red'>5 Puntos</font>) Implemente la función `cases_by_country_plus(cases, countries, date)`, la cual genera un nuevo dataset con los casos para la fecha `date`, con las columnas: `name`,	`region`, `subregion`,	`population` y `total_cases`.



````
def cases_by_country_plus(cases, countries, date):
  """
  Genera un nuevo dataset con los casos para la fecha date, con las columnas: name, region, subregion, population, total_cases.

  Parametros
  ==========
  cases : pd.DataFrame
    Dataset de casos totales
  countries : pd.DataFrame
    Dataset de paises
  date : str
    Fecha en formato 'yyyy-mm-dd'

  Return
  ======
  pd.DataFrame
    Nuevo dataset generado.
  """

  # COMPLETAR ...

  cases_by_date = cases.loc[date]
  cases_by_country = countries.copy()
  countries_names = countries.name.array
  cases_by_country["total_cases"] = cases_by_date[countries_names].values

  return cases_by_country

cases_by_country_plus(cases, countries, "2020-04-14")
````

````
window=14

cases_5_diff = cases_5.diff().rolling(window).mean()

plt.figure(figsize=(15,6))
cases_5_diff.plot(logy=True, ax=plt.gca())
plt.title(f"Casos vs Tiempo");
plt.ylabel("Log Casos")
plt.xlabel("Fecha")
plt.show()
````

* (<font color='red'> 5 Puntos</font>) El comportamiento de la propagación es independiente del tiempo, es decir, el comportamiento en Febrero (como USA y Corea) debe ser el mismo que en Marzo (como Chile) solo que lleva un mes atrasado.
````
plt.figure(figsize=(15,6))
for country, color in zip(["Chile", "United States", "China", "South Korea", "Italy"], ["blue", "orange", "green", "red", "purple"]):
  plt.loglog(cases_5[country].values, cases_5_diff[country].values, label=country, color=color)

plt.title("Crecimiento Exponencial de Sars-CoV-2")
plt.xlabel("Casos Totales")
plt.ylabel("Casos Nuevos")
plt.legend();
````

##4. Mapas
Una de las visualizaciones más interesantes y poderosas que se pueden tener con datos geográficos son las que involucran mapas, ya que entregan mucha información que no se puede obtener de forma sencilla en gráficos de barras u otros. A continuación, veremos 2 tipos de vizualiciones en mapas usando Plotly, y como animar estas vizualiaciones.

Nuevamente, es necesario tener un dataframe en formato Tidy, pero ahora además es necesario incorporar la información de cada país. Para esto, se usa el siguiente código:

````
tidy_cases = cases[countries['name']].reset_index(level=0).melt(id_vars='date',
                                                                var_name='country',
                                                                value_name='total_cases')
dicc = countries.set_index('name').to_dict()
tidy_cases['population'] = [dicc['population'][n] for n in tidy_cases['country']] 
tidy_cases['region'] = [dicc['region'][n] for n in tidy_cases['country']] 
tidy_cases['country_code'] = [dicc['country_code'][n] for n in tidy_cases['country']]
#fechas se transforma de datetime a string para ser graficadas
tidy_cases['date'] = [str(date)[:10] for date in tidy_cases['date']]
tidy_cases.head()
````

###4.1 Scatter Geo
En general un gráfico *scatter* consiste en dibujar los datos como puntos sobre los ejes. En este caso, los ejes son algún mapa geográfico. Este tipo de gráfico también se conoce como BubbleMap.
* (<font color='red'> 3 Puntos</font>) Usando el método scatter_geo, gráfique la cantidad de casos confirmados para todos los países para **una fecha específica**. Utilice el argumento *size* para indicar la cantidad, *locations* para el código del país (esto lo posiciona en el mapa), y puede llenar otros argumentos como *color* o *hover_name*.

````
fig3 = px.scatter_geo(tidy_cases[tidy_cases['date'] == '2020-04-14'],
                     locations='country_code', size='total_cases',
                     projection='orthographic', color='region',
                     hover_name='country')
fig3.show()
````

* (<font color='red'> 3 Puntos</font>) Nuevamente, usando el método scatter_geo, gráfique la cantidad de casos confirmados para todos los países usando **todas las fechas** como una animación. El argumento *animation_frame* se encargará de esto sin problemas.

````
fig4 = px.scatter_geo(tidy_cases, animation_frame='date',
                     locations='country_code', size='total_cases',
                     projection='orthographic', color='region',
                     hover_name='country')
fig4.show()
````

* (<font color='red'> 1 Puntos</font>) Prueben variando el parámetro de proyección, y algún otro que le parezca interezante.



````
fig5 = px.scatter_geo(tidy_cases, locations='country_code', size='total_cases',
                      projection='natural earth', color='region',
                      hover_name='country', animation_frame='date')
fig5.show()
````

###4.2 Choropleth
Un mapa coroplético está compuesto por polígonos de colores. Se usa para representar variaciones de cantidades en el espacio, perfecto para nuestro problema.
* (<font color='red'> 4 Puntos</font>) Usando el método choropleth, gráfique la cantidad de casos confirmados para todos los países para **una fecha específica**. Utilice el argumento *color* para indicar la cantidad, *locations* para el código del país (esto lo posiciona en el mapa). Use el argumento *range_color* para dejar el rango fijo en el intervalo \[min(casos confirmados); max(casos confirmados)\].


````
fig6 = px.choropleth(tidy_cases[tidy_cases['date'] == "2020-04-14"],
                    locations='country_code', color='total_cases',
                    color_continuous_scale='Reds',
                    hover_name='country',
                    range_color=[min(tidy_cases['total_cases']),
                                 max(tidy_cases['total_cases'])])
fig6.show()
````
