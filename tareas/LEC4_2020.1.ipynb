### Verosimilitud en Normal

Ahora repetiremos el procedimiento anterior para una distribución con más de 1 parámetro, en este caso, la Normal. Esta función considera un muestra que denotaremos como $X: \{x_i | \; \forall i \in [0, n]\}$ y parámetros a estimar $\mu$ y $\sigma$.

Sea la función: $$L(X| \mu , \sigma ) = \prod_{i=1}^n P(x_i|\mu , \sigma)$$
$$L(X| \mu , \sigma ) = \prod_{i=1}^n \frac{1}{\sigma \sqrt{2\pi}} e^{\frac{-(x_i - \mu)^2}{2\sigma^2}}$$
$L$ es la función de verosimilitud asociada a la distribución Normal. Obtendremos también la función de log-verosimilitud:
$$\mathcal{L}(X| p ) = - \sum_{i=1}^n  ln(\sigma \sqrt{2\pi}) + \frac{(x_i - \mu)^2}{2\sigma^2}$$

````
from math import log, sqrt, pi
from scipy.stats import norm
from scipy.optimize import minimize


def norm_likelihood(x, params):
  mu = params[0]
  sigma = params[1]
  arg1 = 1 / (sigma * sqrt(2*pi))
  arg2 = np.exp( -1*( (x - mu)**2) / (2*(sigma**2)) )
  return np.prod([arg1 * arg2])

def norm_log_likelihood(x, params):
  mu = params[0]
  sigma = params[1]
  arg1 = -1 * log(sigma * sqrt(2*pi))
  arg2 = -1*( (x - mu)**2) / (2*sigma**2)
  return np.sum([arg1 + arg2])

def minus_norm_likelihood(params, x):
  return -1*norm_likelihood(x, params)

def minus_norm_log_likelihood(params, x):
  return -1*norm_log_likelihood(x, params)

def norm_MLE(x):
  return np.array([np.mean(x), sqrt(np.var(x))])


mu=50
sigma=10
x = norm.rvs(size=1000, loc=mu, scale=sigma)

print('Verosimilitud:', norm_log_likelihood(x, [mu, sigma]))
print('Log-Verosimilitud:', norm_log_likelihood(x, [0, 1]))

min_like = minimize(minus_norm_likelihood, x0=[0, 1], args=x, bounds=[(0, 100), (1, 100)])
min_log = minimize(minus_norm_log_likelihood, x0=[0, 1], args=[x], bounds=[(0, 100), (1, 100)])
print('Máximo para verosimilitud:', min_like.x)
print('Máximo para log-verosimilitud:', min_log.x)

lik = []
log_l = []
mles = []

for n in range(1, 10**4, 10):
  x_aux = norm.rvs(size=n, loc=mu, scale=sigma)

  lik.append(minimize(minus_norm_likelihood, x0=[0, 1], args=x_aux, bounds=[(0, 100), (1, 100)]).x)
  log_l.append(minimize(minus_norm_log_likelihood, x0=[0, 1], args=x_aux, bounds=[(0, 100), (1, 100)]).x)
  mles.append(norm_MLE(x_aux))

plt.title('Estimación de parámetro mu')
plt.plot(range(1, 10**4, 10), np.array(lik)[:, 0], 'blue', label='likelihood', alpha=0.3)
plt.plot(range(1, 10**4, 10), np.array(log_l)[:, 0], 'red', label='log-likelihood', alpha=0.3)
plt.plot(range(1, 10**4, 10), np.array(mles)[:, 0], 'green', label='MLE', alpha=0.3)
plt.legend()
plt.show()

plt.title('Estimación de parámetro sigma')
plt.plot(range(1, 10**4, 10), np.array(lik)[:, 1], 'blue', label='likelihood', alpha=0.3)
plt.plot(range(1, 10**4, 10), np.array(log_l)[:, 1], 'red', label='log-likelihood', alpha=0.3)
plt.plot(range(1, 10**4, 10), np.array(mles)[:, 1], 'green', label='MLE', alpha=0.3)
plt.legend()
plt.show()

plt.title('Error de estimación en parámetro mu')
plt.plot(range(1, 10**4, 10), np.abs(np.array(mles)[:, 0] - np.array(log_l)[:,0]), 'blue')
plt.show()

plt.title('Error de estimación en parámetro sigma')
plt.plot(range(1, 10**4, 10), np.abs(np.array(mles)[:, 1] - np.array(log_l)[:,1]), 'blue')
plt.show()
````

# Regresion Lineal Simple

El objetivo de un modelo de regresión es tratar de explicar la relación que existe entre una
variable dependiente $Y$ con un vector de variables independiente $\vec{X}$.

En un modelo de **regresión lineal simple** tratamos de explicar la relación que existe entre la
variable respuesta $Y$ y una única variable explicativa $X$. Donde se supone la relación entre estas variables es lineal:

$$Y=\beta_0 + \beta_1 X$$

Mediante las técnicas de regresión de una variable $Y$ sobre una variable $X$, buscamos una función que sea una buena aproximación de una muestra de $n$ pares de puntos ($x_i$,$y_i$). 

Esta función que modela los datos es:

$$y_i= \beta_0 + \beta_1 x_i + e_i$$

Donde $\beta_0$ y $\beta_1$ son el intercepto y la pendiente reales; $e_i$ es el error aleatorio obtenido al momento obtener la muestra (error en la medición, error humano, etc).

Nuestro modelo  es $\hat{y_i}$, es una aproximación de $y_i$:

$$\hat{y_i} = \hat{\beta_0} + \hat{\beta_1} x_i$$

 $\hat{\beta_0}$ y $\hat{\beta_1}$ son los parámetros que definen nuestro modelo.

## Intervalos de Confianza

Los intervalos de confianza para $\beta_0$ y $\beta_1$, están definidos por:

\begin{align}
IC(\beta_0) &= \hat\beta_0 	\pm t_{1-\frac{a}{2};n-2}\sqrt{Var(\beta_0 )}\\
IC(\beta_1) &= \hat\beta_1 	\pm t_{1-\frac{a}{2};n-2}\sqrt{Var(\beta_1 )}
\end{align}

Donde:

\begin{align}
\hat{\sigma}^2 &= \frac{1}{n-2}\sum_{i=0}^n(y_i - \hat{y_i})^2\\
Var(\beta_0 )&= \hat{\sigma}^2 \frac{\sum_{i=0}^{n}x_i^2}{n S_{xx}}\\
Var(\beta_1 )&=  \frac{\hat{\sigma}^2}{S_{xx}}
